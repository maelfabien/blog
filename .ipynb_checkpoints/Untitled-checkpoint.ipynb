```python
import glob
```

```python
for post in glob.glob("_posts2/*.md"):
    
    f = open(post, "r")
    content = f.read()
    header = content.split("---")[1]
    
    header = header.replace("layout: single", "layout:post")
    header = header.replace("published: true", "").replace("published: true", "").replace("author_profile: true", "")
    header = header.replace("""comments : true
toc: true
toc_sticky: true
sidebar:
    nav: sidebar-sample""", "")
    header = header.replace("excerpt", "subtitle").replace("categories", "tags")
    print(header.strip())
```
